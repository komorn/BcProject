```python
import pandas as pd
```

```python
names = pd.read_csv("./namesNAT.csv",index_col='protein',sep=',')
taxa = pd.read_csv("./taxaNAT.csv",index_col='protein',sep=',')
seq = pd.read_csv("./sequencesNAT.csv",index_col='protein',sep=',')
```

```python
taxa = taxa.fillna("MISSING VALUE")
```

```python
merged_df = pd.merge(names, taxa, left_on='protein', right_index=True)
```

```python
merged_df_withSeq = pd.merge(merged_df, seq, left_on='protein', right_index=True)
```

```python
merged_df_withSeq.head()
```

```python
merged_df_withSeq['kingdom_name'] = merged_df_withSeq['kingdom_name'].str.replace("Metazoa","Animalia")
merged_df_withSeq['kingdom_name'] = merged_df_withSeq['kingdom_name'].str.replace("Viridiplantae","Plantae")
mask = (merged_df_withSeq['kingdom_name'] == 'MISSING VALUE') & (merged_df_withSeq['superkingdom_name'] == 'Bacteria')
merged_df_withSeq.loc[mask, 'kingdom_name'] = 'Bacteria'
```

```python
merged_df_withSeq.head(20)
```

```python
merged_df_withSeq.drop(columns='full_recommended_name',inplace=True)
merged_df_withSeq.drop(columns='superkingdom_name',inplace=True)
```

```python
merged_df_withSeq.head()
```

```python
pfams = pd.read_csv("./pfams.csv",index_col='protein',sep=',')
```

```python
cypsALL = pd.merge(merged_df_withSeq, pfams, left_on='protein', right_index=True)
```

```python
cypsALL
```

```python
cypsALL['length'] = cypsALL['sequence'].apply(len)
```

```python
cypsALL.to_csv('allCYPs.csv', index=False)
```
