```python
import pandas as pd
```

```python
names= pd.read_csv("./namesNAT.csv")
names = names.fillna("MISSING VALUE")
cytochromes_submit = names[names["full_submitted_name"].str.contains("Cytochrome P450")]
cytochromes_recommend =names[names["full_recommended_name"].str.contains('Cytochrome P450')]
cyps = pd.concat([cytochromes_submit, cytochromes_recommend])
print(cyps.shape)
cyps.head(20)
```

```python
cyps.to_csv('cyps.csv', index=False)
```

```python
sequences = pd.read_csv("./sequencesNAT.csv")
taxa = pd.read_csv("./taxaNAT.csv",index_col='protein',sep=',')
taxa = taxa.fillna("MISSING VALUE")
pfams = pd.read_csv("pfams.csv")
```

```python
sequences.head(20)
```

```python
#merge cyps.csv, sequences.csv, taxa.csv, pfams.csv
namesseq = pd.merge(cyps, sequences, on=['protein'])
nst = pd.merge(namesseq, taxa, on=['protein'])
dfcyps = pd.merge(nst, pfams, on=['protein'])
dfcyps.head(20)
dfcyps.to_csv('dfcyps.csv', index=False)
#add lenghts to this df
```
